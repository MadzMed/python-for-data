## Gauss: Normal Distribution

### setup density function

```python
domain = range(-100,100)
mu = 0 # average
sigma = 5 # variance

f = lambda x : 1/(sqrt(2*pi*pow(sigma,2))) * exp(-pow((x-mu),2)/(2*pow(sigma,2)))
```

### plot

```python
y = [f(x) for x in domain]
plot = plt.plot(domain, y)
```

## Generate Samples

### setup

```python
matrix = np.random.rand(100,200)

sums = np.sum(matrix,0) # sum function from numpy permit different type of sum, in this case, the second argument 
# represent the sum of each line of the matrix. for sum of each column use 1, and null for all the elements

print("sums variables length is {}.".format(sums.shape)) # shape is a function that permit to see size of sums,
# usefull with error ValueError: shapes not aligned
# result is a tuple, a particularity of numpy, permits to read results as line and as column
```

### plot

```python
plot = plt.scatter(range(200), sums) # range 200 due to shape of sums
# result not very usefull with cloud diagram
```

```python
plot = plt.hist(sums, bins= 100) # using histogram is better to see this results
# bins is an argument for the number of classes or histogram column to print
```

```python
# retry the same as previously with more sample and individuals
matrix = np.random.rand(1000, 10000)
sums = np.sum(matrix, 0)
plot = plt.hist(sums, bins = 1000)
# with more individuals and sample, the result is near that a normal distribution
```

## samples property

```python
print("empirical average of our distribution is {}."
      .format(np.mean(sums)))
print("empirical variance of our distribution is {}."
      .format(np.var(sums)))
```

```python
print("empirical average of the variable generate by rand func is {}."
      .format(np.mean(np.random.rand(10000))))
print("empirical variance of the variable generate by rand func is {}."
      .format(np.var(np.random.rand(10000))))
```

we can see that the **empirical average of our distribution** is the **sum of all the variables** generated by rand, same result for variance, *this result is due to the independance of each variable*
